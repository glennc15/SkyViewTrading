```python
%load_ext autoreload
%autoreload 2

import datetime

import CBOEFileParser
```

```python
# cboe_file = "/Users/glenn/Documents/SkyViewTrading/data/SPY_201807101615.dat"
cboe_file = "/Users/glenn/Documents/SkyViewTrading/data/QQQ_201807220252.dat"
# cboe_file = "/Users/glenn/Documents/SkyViewTrading/data/USO_201807220303.dat"
risk_free_rate = 0.0296

option_chains = CBOEFileParser.CBOEFileParser().parse_cboe_file(cboe_file) \
                    .remove_options_after(days=60) \
                    .remove_options_before(days=30) \
                    .set_exchange() \
                    .add_greeks(risk_free_rate=risk_free_rate)

option_chains.pp_expirations()
```

```python
print("Current Spot: {}".format(option_chains.underlying_spot))

for expiration, calls, puts in option_chains.items():
    print("Expiration: {}".format(expiration))
    print("Size of the calls with {} is {}".format(expiration, calls.shape[0]))
    print(calls.head())
    print(calls.tail())

    print("Size of the puts with {} is {}".format(expiration, puts.shape[0]))
    print(puts.head())
    print(puts.tail())



```
